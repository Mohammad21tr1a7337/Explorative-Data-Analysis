#### Using Boolean Masking with Multiple Conditions
Combine multiple conditions using logical operators, When dealing with numpy arrays, the logical operators `or`, `and`, and `not` commonly used in standard Python cannot be applied directly. Instead, for operations involving `True` and `False` within numpy arrays, the symbols `&` represent "and", `|` signify "or", and `~` is used for "not".:
* `&` instead of `and`
* `|` instead of `or`
* `~` for negation

Logical conditions can be combined as well. However, it is important to enclose each individual condition within parentheses to ensure that numpy interprets them correctly.


```python
arr = np.array([10, 15, 20, 25, 30])
# Combine conditions
filtered = arr[(arr > 15) & (arr < 30)]
print(filtered)
```

```python
# Negation example
negated = arr[~(arr > 20)]
print(negated)
```

#### Copy
A copy duplicates both the metadata and the data buffer, creating a new array. Changes in the copy do not affect the original array. Use `ndarray.copy()` to force a copy.


```python
arr = np.array([1, 2, 3])
arr, arr.ctypes.data, arr.base
```

```python
copy_arr = arr
copy_arr[0] = 99
print(arr, copy_arr)
copy_arr.ctypes.data, arr.base, copy_arr.base
```

```python
copy_arr = arr[:]
copy_arr[0] = 99
print(arr, copy_arr)
copy_arr.ctypes.data, arr.base, copy_arr.base
```

Since numpy often generates views when you index a vector, any modifications made to these views can affect the vectors tied to other variables. Therefore, itâ€™s crucial to determine whether the object you're handling is a copy or a reference to avoid unintended changes.

```python
arr = np.array([1, 2, 3])
arr, arr.ctypes.data, arr.base
```

```python
copy_arr = arr.copy()
copy_arr[0] = 99
print(arr, copy_arr)
copy_arr.ctypes.data, arr.base, copy_arr.base
```

As we discussed earlier, numpy returns a view of the array when you subset using simple indexing. However, if you use fancy indexing or Boolean subsetting, numpy creates a copy of the data instead of a view. This distinction is critical to understand to prevent unintended changes in your data.

```python
arr = np.array([1, 2, 3])
print(arr, arr.ctypes.data, arr.base)
new_arr = arr[0:3]
print(new_arr, new_arr.ctypes.data, new_arr.base)
```

```python
arr = np.array([1, 2, 3])
new_arr = arr[0:3].view()
new_arr[[0]] = 10
arr, new_arr
```

```python
arr = np.array([1, 2, 3])
print(arr, arr.ctypes.data, arr.base)
new_arr = arr[[0, 1, 2]]
print(new_arr, new_arr.ctypes.data, new_arr.base)
```

```python
arr = np.array([1, 2, 3])
new_arr = arr[[0, 1, 2]]
new_arr[[0]] = 10
arr, new_arr
```

```python
arr = np.array([1, 2, 3])
print(arr, arr.ctypes.data)
new_arr = arr[arr < 4]
print(new_arr, new_arr.ctypes.data)
```

```python
arr = np.array([1, 2, 3])
new_arr = arr[arr < 4]
new_arr[[0]] = 10
arr, new_arr
```

To address this repeat the experiment for 10000 times

```python
# Using simple indexing to create views
import time
start = time.time()
for _ in range(10000):
  first_half = data[:500000]
  second_half = data[500000:]

end = time.time()
duration_with_views = (end - start) / 10_000
print(f"Subsets with views took {duration_with_views*1000:.4f} milliseconds")
```

```python
# Using fancy indexing to create copy
import time
fh = np.arange(500000)
sh = fh = np.arange(500000, 1000000)
start = time.time()
for _ in range(10000):
  first_half = data[fh]
  second_half = data[sh]

end = time.time()
duration_with_copies = (end - start) / 10_000
print(f"Subsets with copies took {duration_with_copies * 1_000:.4f} milliseconds")
```

```python
# Inserting an element into a 1D Array
arr = np.array([10, 20, 30, 40])
print("Original Array:\n", arr)
```

```python
new_arr = np.insert(arr, 2, 25)  # Insert 25 at index 2
print("After Insertion:", new_arr)
```

```python
np.may_share_memory(arr, new_arr)
```

```python
# Inserting Multiple Values in a 1D Array
new_arr = np.insert(arr, [1, 3], [99, 88])  # Insert 99 and 88
print("After Insertion:", new_arr)
```

```python
# Inserting a Row in a 2D Array
arr2D = np.array([[1, 2, 3], [4, 5, 6]])
print("Original Array:\n", arr2D)
```

### Flatten array using `ravel()` and `flatten()`

```python
# Generate a 5Ã—10 array
arr = np.random.randint(10, 40, size=(5, 10))
print("Numpy Array: \n", arr)
print("Shape: ", arr.shape)
```

```python
# Flatten using `flatten`
flatten_arr = arr.flatten()
print("Flatten Array: \n", flatten_arr)
print("Shape: ", flatten_arr.shape)
```

```python
# Flatten using `ravel`
ravel_arr = arr.ravel()
print("Ravel Array: \n", ravel_arr)
print("Shape: ", ravel_arr.shape)
```

```python
#  Update values in the flattened arrays
print("Original Array: \n", arr)

flatten_arr[1] = 99
print("Flatten Array (After Update): \n", flatten_arr)
print("Original Array (After Update using Flatten): \n", arr)
```

```python
ravel_arr[1] = 99
print("Ravel Array (After Update): \n", ravel_arr)
print("Original Array (After Update using Ravel): \n", arr)
```
